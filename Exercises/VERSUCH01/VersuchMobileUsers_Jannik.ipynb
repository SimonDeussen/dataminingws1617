Für das schreiben der Daten aus dem Pandas Dataframe in die SQLite Datenbank ist die Pandas-Methode [to_sql()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_sql.html) zu verwenden. Für jedes .csv File soll in der SQLite-DB eine eigene Tabelle angelegt werden. Als DB-connector soll eine engine-Instanz des _SQLAlchemy_-Pakets mit der Methode create\_engine() angelegt werden. Siehe z.B. [SQLAlchemy Doku](http://docs.sqlalchemy.org/en/latest/core/engines.html).

2. Nachdem alle Tabellen der DB angelegt sind, sollen aus jeder Tabelle die ersten 10 Zeilen mit der Pandas Methode [read_sql_query()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_sql_query.html) abgefragt und angezeigt werden. Ausserdem ist für jede Tabelle die Größe (Anzahl der Zeilen) auszugeben.

5. Wie viele verschiedene devices kommen in beiden dieser Tabellen vor? 


```python
#-*- coding: utf-8 -*-
%matplotlib inline

from sqlalchemy import create_engine
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
import os
import gmaps
import gmaps.datasets

RESOURCE_PATH = '.'
engine = create_engine('sqlite:///data.db')
tables = [{'table_name': 'app_labels', 'filename': 'app_labels.csv'},
          {'table_name': 'label_categories', 'filename': 'label_categories.csv'},
          {'table_name': 'phone_brand_device_model', 'filename': 'phone_brand_device_model.csv'},
          {'table_name': 'events', 'filename': 'events.csv'},
          {'table_name': 'app_events', 'filename': 'app_events.csv'},
          {'table_name': 'gender_age', 'filename': 'gender_age_train.csv'}]
```

```python
brandMapFilePath = os.path.join(RESOURCE_PATH, 'brandMap.txt')

brandMapFile = open(brandMapFilePath, "r")
brandMapList=[x for x in brandMapFile.read().strip().split() if x]
brandMapList.remove(brandMapList[0])
brandMap= dict()
for i in range(0,len(brandMapList),2):
    brandMap.setdefault(brandMapList[i],brandMapList[i+1])

def translateBrand(word):
    if word in brandMap:
        return brandMap[word]
    else:
        return word
```

```python
# Aufgabe 1 - ACHTUNG, der Code erstellt die SQL Tabelle und läuft sehr lange!

for table in tables:
    for chunk in pd.read_csv(os.path.join(RESOURCE_PATH, table['filename']), chunksize=20000, encoding='utf-8'):
        chunk.to_sql(table['table_name'], engine, if_exists='append')
```

```python
# Aufgabe 2

for table in tables:
    print "Die ersten 10 Zeilen der Tabelle %s: " % (table['table_name'])
    print pd.read_sql_query("SELECT * FROM %s LIMIT 10" % (table['table_name']), engine)
    row_count = pd.read_sql_query("SELECT count(*) FROM %s" % (table['table_name']), engine).at[0, 'count(*)']
    print "Anzahl der Zeilen der Tabelle %s: " % (table['table_name']) + str(row_count)
```

```python
# Aufgabe 3 bis 5
device_id_gender_age = pd.read_sql_query("SELECT DISTINCT device_id FROM gender_age", engine)
print "Anzahl verschiedene devices in der Tabelle gender_age_train: %s" % len(device_id_gender_age)
device_id_events = pd.read_sql_query("SELECT DISTINCT device_id FROM events", engine)
print "Anzahl verschiedener devices in der Tabelle events: %s" % len(device_id_events)
print "Anzahl verschiedener devices in beiden Tabellen: %s" % len(device_id_gender_age.merge(device_id_events, on='device_id', how='outer'))
```

```python
# Read the events.csv
events = pd.read_csv(os.path.join(RESOURCE_PATH, 'events.csv'))
```

```python
# Aufgabe 1

# Remove events that have a zero as longitude and latitude
events_no_zeros = events[(events['longitude'] != 0) | (events['latitude'] != 0)]

# Get the count of the events for every device
device_counts = events_no_zeros['device_id'].value_counts().to_frame(name='deviceCount')

# Filter out the device ids with less than 30 events
device_counts = device_counts[device_counts['deviceCount'] >= 30]

# Get a device id with 30 events or more
deviceId = device_counts.iloc[[1001]].index[0]

# Query the rows from the original events
locations = events_no_zeros[events_no_zeros['device_id'] == deviceId].copy(deep=True)

# Aufgabe 2

locations_matrix =locations.as_matrix(columns=['latitude', 'longitude'])

gmaps.configure(api_key="AIzaSyA_XwCSUXvDQK9EKjLBu4XasqJzlTM74-s")
m = gmaps.Map()
m.add_layer(gmaps.Heatmap(data=locations_matrix.tolist()))
m
```

```python
# Aufgabe 3

epsilon = 0.12
min_samples = 5
db = DBSCAN(eps=epsilon, min_samples=min_samples).fit(locations_matrix)

# Array of all cluster labels
labels = db.labels_

# Array of unique cluster labels
unique_labels = set(labels)

# Number of clusters found
n_clusters = len(unique_labels)

# Add the cluster labels to the locations dataframe
locations['cluster'] = labels

# Assign a new color to every unique cluster
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

# Create a helper function that returns the same colors for each cluster
colorMap = {key: value for (key, value) in zip(unique_labels, colors)}
def getClustorColor(label):
    if label in colorMap:
        return colorMap[label]
    else:
        return 'b'
    

# Set up patches for the legend
patches = [mpatches.Patch(color=getClustorColor(label), label=(label if label >= 0 else 'outliers')) for label in unique_labels] 

# Increase figure size
plt.figure(figsize=(10, 10))

for index, row in locations.iterrows():
    plt.plot(row['latitude'], row['longitude'], 'o', markerfacecolor=getClustorColor(row['cluster']), markersize=14)

plt.legend(handles=patches, loc='best', prop={'size':10})
plt.title('Estimated number of clusters: %d' % n_clusters)
plt.axis([39.5,40,115.5,117])
plt.show()
```

```python
# Aufgabe 4

plt.figure(figsize=(20, 10))

# Convert the timestamp index to datetime
locations['timestamp'] = pd.to_datetime(locations.timestamp)

# Sort the dataframe by date
sorted_locations = locations.sort_values(by='timestamp')

# Draw the line
plt.plot(sorted_locations['timestamp'], range(len(sorted_locations)), 'k-')

# Draw the dots
count = 0
for index, row in sorted_locations.iterrows():
    plt.plot(row['timestamp'], count, 'bo-', markerfacecolor=getClustorColor(row['cluster']))
    count = count + 1
    
plt.title('Smartphone Events over time')
plt.legend(handles=patches, loc='best', prop={'size':15})

x = plt.show()
```
