```python
#Initialisierung
#-*- coding: utf-8 -*-

%matplotlib inline 

from sqlalchemy import create_engine
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
import matplotlib.mlab as mlab
import os
import gmaps
import gmaps.datasets

RESOURCE_PATH = ''

DATA_PATH = RESOURCE_PATH
CHUNK_SIZE = 20000
#Hässlich, aber das haben wir nicht richtig abgeklärt

engine = create_engine('sqlite:///data.db')
tables = [{'table_name': 'app_labels', 'filename': 'app_labels.csv'},
          {'table_name': 'label_categories', 'filename': 'label_categories.csv'},
          {'table_name': 'phone_brand_device_model', 'filename': 'phone_brand_device_model.csv'},
          {'table_name': 'events', 'filename': 'events.csv'},
          {'table_name': 'app_events', 'filename': 'app_events.csv'},
          {'table_name': 'gender_age', 'filename': 'gender_age_train.csv'}]


raw_data_app_events = pd.read_csv(DATA_PATH + "app_events.csv")
raw_data_app_labels = pd.read_csv(DATA_PATH + "app_labels.csv")
raw_data_events = pd.read_csv(DATA_PATH + "events.csv")
raw_data_gender_age_test = pd.read_csv(DATA_PATH + "gender_age_test.csv")
raw_data_gender_age_train = pd.read_csv(DATA_PATH + "gender_age_train.csv")
raw_data_phone_brand_device_model = pd.read_csv(DATA_PATH + "phone_brand_device_model.csv")
raw_data_sample_submission = pd.read_csv(DATA_PATH + "sample_submission.csv")
raw_data_label_categories = pd.read_csv(DATA_PATH + "label_categories.csv")

raw_data = [raw_data_app_events, raw_data_app_labels, raw_data_events, 
            raw_data_gender_age_test, raw_data_gender_age_train, 
            raw_data_phone_brand_device_model, raw_data_sample_submission, raw_data_label_categories]
raw_data_labels = { 0:"app_events", 1:"app_labels", 2:"events", 3:"gender_age_test", 
                  4:"gender_age_train", 5:"phone_brand_device_model", 6:"sample_submission", 7:"label_categories" }

def gln(S):
    for i in range(0,len(raw_data_labels)):
        if raw_data_labels[i] == S:
            return raw_data[i]
    print "NOT FOUND"
    return
```

```python
# Funktion zum Übersetzen der chinesischen Markennamen

brandMapFilePath = os.path.join(RESOURCE_PATH, 'brandMap.txt')

brandMapFile = open(brandMapFilePath, "r")
brandMapList=[x for x in brandMapFile.read().strip().split() if x]
brandMapList.remove(brandMapList[0])
brandMap= dict()
for i in range(0,len(brandMapList),2):
    brandMap.setdefault(brandMapList[i],brandMapList[i+1])

def translateBrand(word):
    if word in brandMap:
        return brandMap[word]
    else:
        return word
```

```python
# Aufgabe 1 - ACHTUNG, der Code erstellt die SQL Tabelle und läuft sehr lange!

for table in tables:
    for chunk in pd.read_csv(os.path.join(RESOURCE_PATH, table['filename']), chunksize=20000, encoding='utf-8'):
        chunk.to_sql(table['table_name'], engine, if_exists='append')
```

```python


# Remove events that have a zero as longitude and latitude
events_no_zeros = events[(events['longitude'] != 0) | (events['latitude'] != 0)]

# Get the count of the events for every device
device_counts = events_no_zeros['device_id'].value_counts().to_frame(name='deviceCount')

# Filter out the device ids with less than 30 events
device_counts = device_counts[device_counts['deviceCount'] >= 30]

# Get a device id with 30 events or more
deviceId = device_counts.iloc[[1001]].index[0]

# Query the rows from the original events
locations = events_no_zeros[events_no_zeros['device_id'] == deviceId].copy(deep=True)

# Aufgabe 2

locations_matrix =locations.as_matrix(columns=['latitude', 'longitude'])

gmaps.configure(api_key="AIzaSyA_XwCSUXvDQK9EKjLBu4XasqJzlTM74-s")
m = gmaps.Map()
m.add_layer(gmaps.Heatmap(data=locations_matrix.tolist()))
m
```

```python
# Aufgabe 3

epsilon = 0.12
min_samples = 5
db = DBSCAN(eps=epsilon, min_samples=min_samples).fit(locations_matrix)

# Array of all cluster labels
labels = db.labels_

# Array of unique cluster labels
unique_labels = set(labels)

# Number of clusters found
n_clusters = len(unique_labels)

# Add the cluster labels to the locations dataframe
locations['cluster'] = labels

# Assign a new color to every unique cluster
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

# Create a helper function that returns the same colors for each cluster
colorMap = {key: value for (key, value) in zip(unique_labels, colors)}
def getClustorColor(label):
    if label in colorMap:
        return colorMap[label]
    else:
        return 'b'
    

# Set up patches for the legend
patches = [mpatches.Patch(color=getClustorColor(label), label=(label if label >= 0 else 'outliers')) for label in unique_labels] 

# Increase figure size
plt.figure(figsize=(10, 10))

for index, row in locations.iterrows():
    plt.plot(row['latitude'], row['longitude'], 'o', markerfacecolor=getClustorColor(row['cluster']), markersize=14)

plt.legend(handles=patches, loc='best', prop={'size':10})
plt.title('Estimated number of clusters: %d' % n_clusters)
plt.axis([39.5,40,115.5,117])
plt.show()
```

```python
#takes dataframe and returns a random dataset by id
def getRandomDataset(dataframe):
    #get all IDs, make a list
    idList = dataframe[dataframe.columns[0]].unique()
    #get a random id from list
    randomID = idList[random.randint(0,len(idList)-1)]
    #get the random dataset by id
    randomDataSet = dataframe.loc[dataframe[dataframe.columns[0]] == randomID]
   
    return randomDataSet
```

```python
#takes dataframe and returns python dictionary with unique values and headers
def createMatrixSizeDic(dataframe):
    #slice our id away
    dataframe = dataframe.drop(dataframe.columns[0], axis=1)
    dataframe = dataframe.drop('group', axis=1)
    #get headers to list
    headers = list(dataframe.columns.values)
    #count values distinct
    sizeDic = {headers[x]:dataframe[headers[x]].unique() for x in range(len(headers))}
    
    return sizeDic
```

```python
#creates a one hot encode array
def createMainDataOneHotEncode(dataframeall, dataframe):
    oneHotEncodeArray = []
    
    #get all IDs, make a list
    idList = dataframe[dataframe.columns[0]].unique()
    #get the dataframe size Dictionary
    sizeDic = createMatrixSizeDic(dataframeall)

    for ind in range(len(idList)):
        
        #get the random dataset by id
        DataSet = dataframe.loc[dataframe[dataframe.columns[0]] == idList[ind]]
        
        #get random Group
        #DataSetGroup = DataSet['group'].unique()
        #DataSetGroupStr = randomDataSetGroup[0]
    
#    print randomDataSetGroupStr

        #drop the id row of the random Dataset
        DataSet = DataSet.drop(DataSet.columns[0], axis=1)
    
        #drop group Row
        DataSet = DataSet.drop('group', axis=1)
    
        #get headers of randomDataSet
        headers = list(DataSet.columns.values)
    
        #create utilityLists oneHotEncodeDataset with sizeDic Parameters and randomDataset
        tmpList = []
        for key, value in sizeDic.iteritems():
            for x in value:
                tmpList.append(x)
            
#   print tmpList
    
        tmpList2 = []
        for column, row in DataSet.iterrows():
            for i in range(len(headers)):
                tmpList2.append(row[headers[i]])
            
        #remove duplicates
        tmpList2 = set(tmpList2)
        
#   print tmpList2
    
        #replace in tmpList all values with 1 when found in tmpList2 the rest values with 0
        for index, item in enumerate(tmpList2):
            for ind, val in enumerate(tmpList):
                if (item == val):
                    tmpList[ind] = 1
    
        for index, item in enumerate(tmpList):
            if (item != 1):
                tmpList[index] = 0
            
        oneHotEncodeArray.extend(tmpList)
        
        #for z in range(134):
        #    zero = [0]
        #    oneHotEncodeArray.extend(zero)
#   print len(oneHotEncodeArray[randomDataSetGroupStr:tmpList])
    
    return oneHotEncodeArray
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
#cnf_matrix = confusion_matrix(y_test, y_pred)

```

```python
import numpy as np

def multiclass_log_loss(y_true, y_pred, eps=1e-15):
    """Multi class version of Logarithmic Loss metric.
    https://www.kaggle.com/wiki/MultiClassLogLoss

    idea from this post:
    http://www.kaggle.com/c/emc-data-science/forums/t/2149/is-anyone-noticing-difference-betwen-validation-and-leaderboard-error/12209#post12209

    Parameters
    ----------
    y_true : array, shape = [n_samples]
    y_pred : array, shape = [n_samples, n_classes]

    Returns
    -------
    loss : float
    """
    predictions = np.clip(y_pred, eps, 1 - eps)

    # normalize row sums to 1
    predictions /= predictions.sum(axis=1)[:, np.newaxis]

    actual = np.zeros(y_pred.shape)
    rows = actual.shape[0]
    actual[np.arange(rows), y_true.astype(int)] = 1
    vsota = np.sum(actual * np.log(predictions))
    return -1.0 / rows * vsota
```
